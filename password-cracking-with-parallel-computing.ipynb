```python
from concurrent.futures import ProcessPoolExecutor, ThreadPoolExecutor, as_completed
import itertools
import string
import logging
```

```python
logging.basicConfig(level=logging.DEBUG, format="%(levelname)s: %(message)s")
```

```python
cores = 8
target_string = "$2y$10$N86jMDU"  
```

```python
def attempt_match(length, target_string):
    try:
       
        logging.debug(f"Starting attempt_match with length {length}")
        
        for attempt in itertools.product(string.ascii_lowercase, repeat=length):
            attempt = ''.join(attempt)
            if attempt == target_string:
                logging.info(f"Password found: {attempt}")
                return attempt
        return None  
    except Exception as e:
        logging.error(f"Error in process for length {length}: {e}")
        return None
```

```python
with ThreadPoolExecutor(max_workers=cores) as executor:
    futures = [executor.submit(attempt_match, length, target_string) for length in range(1, len(target_string)+1)]
    for future in as_completed(futures):
        match = future.result()
        if match:
            print("Matching password:", match)
            break
    else:
        print("No match found.")
```
